```python
import tensorflow as tf
```

```python
import numpy as np
```

```python
file = 'images//pikachu.jpg'
```

```python
from tensorflow.keras.preprocessing import image
```

```python
import matplotlib.pyplot as plt
```

```python
img1 = image.load_img(file, target_size=(224,224))
plt.imshow(img1)
```

```python
resized_img = image.img_to_array(img1)
final_image = np.expand_dims(resized_img, axis=0)
final_image = tf.keras.applications.mobilenet.preprocess_input(final_image)
```

```python
final_image.shape
```

```python
mobile = tf.keras.applications.mobilenet.MobileNet()
predictions = mobile.predict(final_image)
```

```python
from tensorflow.keras.applications import imagenet_utils
```

```python
results = imagenet_utils.decode_predictions(predictions)
```

```python
print(results)
```
